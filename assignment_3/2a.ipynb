```python
# Load the CSV files
train_data = pd.read_csv('./data/fashion-mnist_train.csv')
test_data = pd.read_csv('./data/fashion-mnist_test.csv')

X_train = train_data.iloc[:, 1:].values  
y_train = train_data.iloc[:, 0].values  

X_test = test_data.iloc[:, 1:].values    
y_test = test_data.iloc[:, 0].values  


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Summary statistics for pixel values
train_mean = np.mean(X_train)
train_std = np.std(X_train)
train_min = np.min(X_train)
train_max = np.max(X_train)

# Class distribution
train_labels, train_counts = np.unique(y_train, return_counts=True)
class_distribution = dict(zip(train_labels, train_counts))

# Printing the summary
print("Training Set Statistics:")
print(f"Number of samples: {X_train.shape[0]}")
print(f"Number of features: {X_train.shape[1]}")
print(f"Mean pixel value: {train_mean:.2f}")
print(f"Standard deviation of pixel values: {train_std:.2f}")
print(f"Minimum pixel value: {train_min}")
print(f"Maximum pixel value: {train_max}")
print()

print("Class Distribution (Training Set):")
for label, count in class_distribution.items():
    print(f"Class {label}: {count} samples")

# Visualizing class distribution
plt.figure(figsize=(10, 5))
plt.bar(class_distribution.keys(), class_distribution.values(), color='darkseagreen')
plt.title("Class Distribution in Fashion MNIST Training Set")
plt.xlabel("Class Labels")
plt.ylabel("Number of Samples")
plt.xticks(ticks=list(class_distribution.keys()))
plt.grid(axis='y')
plt.show()
```
